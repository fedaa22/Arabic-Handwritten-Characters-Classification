```python
import os
import cv2
import numpy as np

IMG_SIZE = 64

def load_dataset(dataset_path):
    X = []  # Store images
    y = []  # Store labels

    # Get class names (folder names) in alphabetical order
    classes = sorted(os.listdir(dataset_path))

    # Map each class name to a numeric ID (0, 1, 2, ...)
    label_map = {
        class_name: idx
        for idx, class_name in enumerate(classes)
    }

    print(f"There are {len(classes)} different classes to work with.")

    # Loop through each class folder
    for class_name in classes:
        class_path = os.path.join(dataset_path, class_name)

        # Skip if it's a file (just in case)
        if not os.path.isdir(class_path):
            continue

        label = label_map[class_name]

        # Loop through every image in the class folder
        for img_name in os.listdir(class_path):
            img_path = os.path.join(class_path, img_name)

            try:
                # Read image in grayscale mode
                img = cv2.imread(
                    img_path,
                    cv2.IMREAD_GRAYSCALE
                )

                # Skip if image failed to load
                if img is None:
                    continue

                # Resize all images to the same fixed size
                img = cv2.resize(
                    img,
                    (IMG_SIZE, IMG_SIZE)
                )

                # Normalize pixel values from [0, 255] to [0, 1]
                img = img / 255.0

                X.append(img)
                y.append(label)

            except Exception as e:
                print(f"Oops! Had some trouble loading this one: {img_path}")
                print(f"Technical details: {e}")

    # Convert lists to numpy arrays
    X = np.array(X)
    y = np.array(y)

    # Reshape for CNN: add channel dimension (grayscale = 1 channel)
    X = X.reshape(
        -1,
        IMG_SIZE,
        IMG_SIZE,
        1
    )

    print("\nDataset is all loaded and ready to go!")
    print(f"-> Final image tensor shape: {X.shape}")
    print(f"-> Final labels array shape: {y.shape}")

    return X, y, label_map
```
